```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
feature = {}
for k in df.columns:
    if df[k].dtypes=='O':
        df[str(k)+'_Encode'] = lb.fit_transform(df[k])
        gt = {}
        ke = df[[k,str(k)+'_Encode']].drop_duplicates()[str(k)+'_Encode'].values
        val = df[[k,str(k)+'_Encode']].drop_duplicates()[k].values
        for i in range(len(ke)):
            gt[val[i]] = str(ke[i])
        feature[k] = gt
```

```python
feature
```

```python
import json
```

```python
with open('encdata.json','w') as jsonfile:
    json.dump(feature,jsonfile,indent=4)
```

```python
y = df["Price"]
x = df.drop(["Price","Condition_Encode"], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
